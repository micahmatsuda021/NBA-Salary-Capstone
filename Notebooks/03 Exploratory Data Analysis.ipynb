## 3.6 Feature Engineering

```python
players_data.columns
```

```python
condition = [(players_data['Salary'] < 5000000)]
choice = [0]
players_data['Binary'] = np.select(condition, choice, default=1)
players_data.head(5)
```

```python
condition = [
        (players_data['Salary'] < 1000000),
        (players_data['Salary'] <= 5000000),
        (players_data['Salary'] <= 10000000)]
choice = [0, 1, 2]
players_data['Nominal'] = np.select(condition, choice, default=3)
players_data.head(10)
```

```python
print(players_data.Binary.value_counts())
p = players_data.Binary.value_counts().plot(kind = "bar", color = ("#E69F00", "#7600bc"))
```

```python
print(players_data.Nominal.value_counts())
p = players_data.Nominal.value_counts().plot(kind = "pie")
```

## 3.7 Feature Importance

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import permutation_importance
import shap
from sklearn import preprocessing

plt.rcParams.update({'figure.figsize': (12.0, 8.0)})
plt.rcParams.update({'font.size': 14})
```

```python
position_map = {'C':0, 'PF':1, 'SF':2, 'PG':3, 'SG':4, 'PF-C':5, 'SG-PG':6, 'SF-PF':7, 'SG-SF':8, 'PG-SG':9}
players_data['Pos'] = players_data['Pos'].map(position_map)
```

```python
players_data.info()
```

```python
X = players_data.drop(['Player', 'Ht', 'Wt', 'Colleges', 'Stats_Year', 'Tm', 'Team', 'Contract_Team', 'Contract_Year', 'Salary', 'Salary Cap', 'age_bin', 'Adjusted Salary 2022', 'Nominal', 'Binary'], axis=1)
y = players_data['Adjusted Salary 2022']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=12)
```

```python
rf = RandomForestRegressor(n_estimators=100)
rf.fit(X_train, y_train)
```

```python
rf.feature_importances_
```

```python
explainer = shap.TreeExplainer(rf)
shap_values = explainer.shap_values(X_test)
```

```python
shap.summary_plot(shap_values, X_test, plot_type="bar")
```
